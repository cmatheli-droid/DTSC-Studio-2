```python
df = df.drop(columns=['Contig36312_RC', 'Contig38980_RC', 'NM_000853', 'NM_000854',
       'NM_000860', 'Contig29014_RC', 'Contig46616_RC', 'NM_000888',
       'NM_000898', 'AF067420'])
```

```python
df.columns
```

```python
df = df[['Patient', 'ID', 'age', 'eventdeath', 'survival', 'timerecurrence',
       'chemo', 'hormonal', 'amputation', 'histtype']]
```

```python
# use seaborn's boxplot which accepts 'x', 'y', 'data' keyword arguments
sns.boxplot(x='eventdeath', y='age', data=df)
plt.xlabel('eventdeath')
plt.ylabel('age')
plt.title('Age distribution by eventdeath')
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
## choose the columns for exact correlation to question (I looked this up with CoPilot)
subset = df[["age", "eventdeath", "survival", "chemo"]]
# show correlation on the numeric subset 
corr = subset.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation matrix (age, eventdeath, survival, chemo)")
plt.show()
```
